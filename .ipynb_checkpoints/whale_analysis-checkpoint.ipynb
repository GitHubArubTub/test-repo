## S&P TSX 60 Returns

Read the S&P TSX 60 historic closing prices and create a new daily returns DataFrame from the data. 

```python
# Reading S&P TSX 60 Closing Prices
csv_path_2 = Path(r'C:\Users\RAMZI\Desktop\BootCamp\Module 4\Starter_Code (1)\Starter_Code\Resources\sp_tsx_history.csv')

# Read the SPYTSX daily returns
tsx_returns = pd.read_csv(csv_path_2) #, dtype={'Close': str})

# Convert date column format
tsx_returns['Date'] = pd.to_datetime(tsx_returns['Date'], format="%d-%b-%Y").dt.strftime('%Y-%m-%d')


# Display the first few rows of the DataFrame
tsx_returns.head()
```

```python
# # Check Data Types

tsx_returns.columns
tsx_returns.dtypes
```

```python
# Fix Data Types
tsx_returns['Close'] = tsx_returns['Close'].astype(str)
tsx_returns['Close'] = tsx_returns['Close'].str.replace(',', '').str.replace('$', '', regex=False)
tsx_returns['Close'] = pd.to_numeric(tsx_returns['Close'])

tsx_returns.dtypes
```

```python
# Calculate Daily Returns

# Calculate daily returns
tsx_returns['Close'] = tsx_returns['Close'].pct_change()

# Display the updated DataFrame
tsx_returns.head()

```

```python
# Drop nulls

tsx_returns = tsx_returns.dropna()

```

```python
# Rename `Close` Column to be specific to this portfolio.

tsx_returns = tsx_returns.rename(columns={'Close': 'S&P TSX %'})

# Display the updated DataFrame
tsx_returns.head()
```

## Combine Whale, Algorithmic, and S&P TSX 60 Returns

```python
# Join Whale Returns, Algorithmic Returns, and the S&P TSX 60 Returns into a single DataFrame with columns for each portfolio's returns.

# Concatenate the DataFrames using inner join
con_returns = pd.concat([whale_returns, algo_returns, tsx_returns], axis=1, join="inner")

# Remove duplicate "Date" columns
con_returns = con_returns.loc[:, ~con_returns.columns.duplicated()]


con_returns = con_returns.dropna()

# Display the resulting DataFrame
con_returns.head()
```

# Sharpe Ratios
In reality, investment managers and thier institutional investors look at the ratio of return-to-risk, and not just returns alone. After all, if you could invest in one of two portfolios, and each offered the same 10% return, yet one offered lower risk, you'd take that one, right?

### Using the daily returns, calculate and visualize the Sharpe ratios using a bar plot

```python
# Annualized Sharpe Ratios
sharpe_ratios = (con_returns.mean() * 252) / (con_returns.std() * (252 ** 0.5))
```

```python
# Visualize the sharpe ratios as a bar plot
sharpe_ratios = sharpe_ratios.drop("Date")

# Create a bar plot of the Sharpe ratios
plt.figure(figsize=(10, 6))
sharpe_ratios.plot(kind='bar')
plt.xlabel('Portfolio')
plt.ylabel('Sharpe Ratio')
plt.title('Sharpe Ratios of Portfolios')
plt.show()
```

### Calculate and Plot the 60-day Rolling Beta for Your Portfolio compared to the S&P 60 TSX

```python
# Calculate and plot Beta

rolling_covariance = port_comb['Cumulative Returns'].rolling(window=60).cov(tsx_returns['S&P TSX %'])

# Calculate rolling variance
rolling_variance = tsx_returns['S&P TSX %'].rolling(window=60).var()

# Calculate rolling beta
rolling_beta = rolling_covariance / rolling_variance

# Plot rolling beta
plt.figure(figsize=(10, 6))
plt.plot(rolling_beta, color='blue')
plt.axhline(y=1.0, color='red', linestyle='--')
plt.title('60-day Rolling Beta')
plt.xlabel('Date')
plt.ylabel('Beta')
plt.legend(['Rolling Beta', 'Baseline Beta'])
plt.grid(True)
plt.show()
```

### Using the daily returns, calculate and visualize the Sharpe ratios using a bar plot

```python
# Calculate Annualized Sharpe Ratios

daily_returns = port_comb.pct_change()

# Calculate average daily return and standard deviation
average_daily_return = daily_returns.mean()
daily_standard_deviation = daily_returns.std()

# Calculate Sharpe ratio
risk_free_rate = 0.02  # Assuming a risk-free rate of 2% (0.02)
sharpe_ratios = (average_daily_return - risk_free_rate) / daily_standard_deviation
annualized_sharpe_ratios = sharpe_ratios * np.sqrt(252)
```

```python
# Visualize the sharpe ratios as a bar plot

plt.figure(figsize=(10, 6))
annualized_sharpe_ratios.plot(kind='bar', color='blue')
plt.title('Annualized Sharpe Ratios')
plt.xlabel('Assets')
plt.ylabel('Sharpe Ratio')
plt.grid(True)
plt.show()
```
